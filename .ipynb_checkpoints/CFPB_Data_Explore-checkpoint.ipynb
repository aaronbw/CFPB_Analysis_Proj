# CFPB Data Explore

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Filter to ignore math related warnings
import warnings
warnings.filterwarnings(action="ignore", message="^internal gelsd")
```

```python
# Initial load of CFPB csv file to explore the data
complaints = pd.read_csv('./complaints.csv')
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
complaints.head()
```

```python
complaints['Company'].value_counts()
```

```python
complaints.info()
```

```python
complaints['Date received'] = pd.to_datetime(complaints['Date received'], yearfirst=True)
```

```python
complaints['Date sent to company'] = pd.to_datetime(complaints['Date sent to company'], yearfirst=True)
```

```python
complaints.info()
```

```python
start_date = '2019-01-01'
end_date = '2020-12-31'
```

```python
mask = (complaints['Date received'] > start_date) & (complaints['Date received'] <= end_date)
```

```python
complaints_2019to2020 = complaints.loc[mask]
complaints_2019to2020.shape
```

```python
companies = ['TRUSTMARK CORPORATION', 'BANCORPSOUTH BANK', 'CADENCE BANCORPORATION', 'RENASANT CORPORATION']
```

```python
companies
```

```python
complaints_2019to2020_4comp = complaints_2019to2020[complaints['Company'].isin(companies)]
complaints_2019to2020_4comp.shape
```

```python
complaints_2019to2020.to_excel('complaints_2019to2020.xlsx')
```

```python
complaints_2019to2020_4comp.to_excel('complaints_2019to2020_4comp.xlsx')
```
